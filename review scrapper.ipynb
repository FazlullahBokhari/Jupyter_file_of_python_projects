```python
 commentboxes = prod_html.find_all('div', {'class': "_16PBlm"})
```

```python
commentboxes
```

```python
len(commentboxes)
```

```python
commentboxes[0].div.div
```

```python
commentboxes[0].div.div.find_all('p', {'class': '_2sc7ZR _2V5EHH'})[0].text
```

```python
name = []
star = []
value_for_money = []
comtag = []
comment = []
for i in range(len(commentboxes)-1):
    name.append(commentboxes[i].div.div.find_all('p', {'class': '_2sc7ZR _2V5EHH'})[0].text)
    star.append(commentboxes[i].div.div.div.div.text)
    value_for_money.append(commentboxes[i].div.div.div.p.text)
    #print(commentboxes[i].div.div.find_all('div', {'class': ''}))
    comtag = commentboxes[i].div.div.find_all('div', {'class': ''})
    comment.append(comtag[0].div.text)
```

```python
import pandas as pd
```

```python
print(len(name),len(star),len(value_for_money),len(comment))
```

```python
data = {'name':name,'star':star,'value_for_money':value_for_money,'comment':comment}
```
